# What do you think about the students' parental level of education

```python
print(' students parental level of education')
df.groupby("parental_level_of_education").count()
```

```python
df['Total_marks']=df.loc[:,'math_score':'writing_score'].sum(axis = 1)
df.head()
```

```python
#Graph showing the parental level of education
```

```python
sns.countplot(x='parental_level_of_education', hue='gender',data=df)
```

```python
sns.boxplot(x='parental_level_of_education',y='Total_marks',data=df)

plt.tight_layout()
plt.show()
```

```python
#Inference on parental level of education
```

```python
# From the above graph it is evident that there is some relationship between parental level of education and total marks scored by students
Total marks secured by students is high among those whose parents are at higher level in education.
```

# Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation cours

```python
#Finding the average score of marks secured by students
```

```python
df['Average_score']=df.loc[:,'math_score':'writing_score'].mean(axis = 1)
df.head(10)
```

```python
df.sort_values(by=['Average_score','gender'],ascending=[False,True]).head(5)
```

```python
df.sort_values(by=['Average_score','test_preparation_course'],ascending=[False,True]).head(5)
```

# What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course

```python
df['standard_deviation']=df.loc[:,'math_score':'writing_score'].std(axis = 1)
df
```

```python
df.plot( x='gender', y='standard_deviation', kind='hist')
```

```python
sns.countplot(x='standard_deviation',hue='gender',data=df )
```

```python
df.plot.scatter(x="standard_deviation", y="gender")
```

```python
#inference: As is seen in the graph the spread of data is very large and the data are loosely clustered in respect of female. So, there is high degree of variation in respect of all three kinds of marks especially among female
```

```python
#finding the variation of marks related to 'test preparation'
```

```python
df.plot.scatter(x="standard_deviation", y="test_preparation_course")
```

```python
# It is observed that the variation of marks is less in respect of those who completed the course than others
```
